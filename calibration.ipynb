```python
import cv2
import datetime
import matplotlib
import matplotlib.pyplot as plt
import imutils
%matplotlib inline
```

```python
DEEPCAM_FFSERVER_URL = 'http://deepcam.local:8090/camera.mjpeg'
def capture_single_image():
    cam = cv2.VideoCapture(DEEPCAM_FFSERVER_URL)
    success, frame = cam.read()
    if not success:
        print('couldnt capture image')
    return frame
cal = capture_single_image()
```

```python
f = plt.figure(figsize=(20,20))

def show_im(n, im, title):
    ax = f.add_subplot(2,2, n)
    ax.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
    ax.set_title(title)

flipped = cv2.flip(cal, -1 ) # flip (camera mounted upside down)
show_im(1, flipped, "original")

ROTATE_ANGLE = -9
rotated = imutils.rotate(flipped, angle=ROTATE_ANGLE) # rotate
show_im(2, rotated, "rotated")

clipped = rotated[200:1200, 500:2000] # clip
show_im(3, clipped, "clipped")

road = clipped[610:750, 180:2000] # clip
show_im(4, road, "road")
```
