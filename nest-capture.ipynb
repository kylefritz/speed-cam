```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
import imutils
import time
%matplotlib inline
import numpy as np
from dotenv import load_dotenv
import os
```

```python
load_dotenv()

m3u8_id = os.getenv('NEST_M3U8_ID')
url = "https://stream-us1-alfa.dropcam.com:443/nexus_aac/{}/playlist.m3u8".format(m3u8_id)
cam = cv2.VideoCapture(url)
success, frame = cam.read()
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
images = []
for x in range(100):
    success, frame = cam.read()
    if success:
        images.append(frame)
        print(x)
        time.sleep(1)
```

```python
f = plt.figure(figsize=(20,100))
for n in range(100):
    ax = f.add_subplot(20,5, n+1)
    im_n = n
    ax.imshow(cv2.cvtColor(images[im_n], cv2.COLOR_BGR2RGB))
    ax.set_title(str(im_n))
```
